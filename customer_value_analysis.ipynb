```python
customer_df["satisfaction_flag"] = np.where(
    customer_df["review_rating"] >= 4,
    "Satisfied",
    "Needs Improvement"
)
```

```python
customer_df["satisfaction_flag"].value_counts()
```

```python
customer_df["promo_dependency"] = np.where(
    (customer_df["discount_applied"] == "Yes") &
    (customer_df["promo_code_used"] == "Yes"),
    "High",
    "Low"
)
```

```python
customer_df["promo_dependency"].value_counts()
```

```python
frequency_map = {
    "Weekly": 5,
    "Fortnightly": 4,
    "Monthly": 3,
    "Every 3 Months": 2,
    "Quarterly": 2,
    "Annually": 1,
    "Bi-Weekly": 4
}

customer_df["purchase_frequency_score"] = (
    customer_df["frequency_of_purchases"]
    .map(frequency_map)
)
```

```python
customer_df[[
    "frequency_of_purchases",
    "purchase_frequency_score"
]].head()
```

```python
customer_df["purchase_amount_norm"] = (
    customer_df["purchase_amount_usd"] -
    customer_df["purchase_amount_usd"].min()
) / (
    customer_df["purchase_amount_usd"].max() -
    customer_df["purchase_amount_usd"].min()
)
```

```python
customer_df["previous_purchases_norm"] = (
    customer_df["previous_purchases"] -
    customer_df["previous_purchases"].min()
) / (
    customer_df["previous_purchases"].max() -
    customer_df["previous_purchases"].min()
)
```

```python
customer_df["frequency_score_norm"] = (
    customer_df["purchase_frequency_score"] -
    customer_df["purchase_frequency_score"].min()
) / (
    customer_df["purchase_frequency_score"].max() -
    customer_df["purchase_frequency_score"].min()
)
```

```python
customer_df["customer_value_score"] = (
    0.4 * customer_df["purchase_amount_norm"] +
    0.4 * customer_df["previous_purchases_norm"] +
    0.2 * customer_df["frequency_score_norm"]
)
```

```python
customer_df["customer_value_score"].describe()
```

```python
customer_df["value_tier"] = pd.qcut(
    customer_df["customer_value_score"],
    q=3,
    labels=["Low", "Medium", "High"]
)
```

```python
customer_df["value_tier"].value_counts()
```

```python
customer_df.to_csv("customer_features.csv", index=False)
```

```python
customer_df["loyalty_a"] = np.where(
    (
        (customer_df["purchase_frequency_score"] >= 4) &
        (customer_df["previous_purchases"] >= customer_df["previous_purchases"].median()) &
        (customer_df["promo_dependency"] == "Low")
    ),
    "Loyal",
    "Not Loyal"
)
```

```python
customer_df["loyalty_a"].value_counts()
```

```python
customer_df["loyalty_b"] = np.where(
    (
        (customer_df["customer_value_score"] >= customer_df["customer_value_score"].median()) &
        (customer_df["satisfaction_flag"] == "Satisfied")
    ),
    "Loyal",
    "Not Loyal"
)
```
